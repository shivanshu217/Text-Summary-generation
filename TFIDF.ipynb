```python
from nltk import word_tokenize,sent_tokenize
from nltk.corpus import stopwords
import string
import math
```

```python
f = open("summarize.txt", "r")
Text = f.read()
```

```python
stop = stopwords.words("english")
stop = stop + list(string.punctuation)
```

```python
def get_TF_IDF(file):
    f = open(file, "r")
    text = f.read()
    arr = sent_tokenize(text)
    new_arr=[]
    total_words=[]
    for w in arr:
        words = word_tokenize(w)
        clean=[]
        for x in words:
            if x.lower() not in stop:
                clean.append(x.lower())
                total_words.append(x.lower())
        new_arr.append(clean)
    #print(new_arr[0])
    dictionary = {}
    for i in range(len(new_arr)):
        dictionary[i]={}
        for x in new_arr[i]:
            if x in dictionary[i].keys():
                dictionary[i][x]+=1
            else:
                dictionary[i][x]=1
    #print(dictionary)
    #Term Frequency
    TF={}
    for i in range(len(new_arr)):
        TF[i]={}
        for x in new_arr[i]:
            TF[i][x]=dictionary[i][x]/len(new_arr[i])
    word_cnt = {}
    total_words = set(total_words)
    for x in total_words:
        word_cnt[x]=0
    for x in total_words:
        for i in range(len(new_arr)):
            if x in new_arr[i]:
                word_cnt[x]+=1

    IDF={}
    for x in total_words:
        IDF[x]=math.log(len(new_arr)/word_cnt[x])
    #print(IDF)
    TF_IDF={}
    for i in range(len(new_arr)):
        TF_IDF[i]={}
        for x in new_arr[i]:
            TF_IDF[i][x]=TF[i][x]*IDF[x]
    return TF_IDF
    #print(TF_IDF)
```
